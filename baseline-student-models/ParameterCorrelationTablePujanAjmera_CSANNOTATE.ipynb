```
import numpy as np
import pandas as pd
import io
import requests
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")    
```

## Parameter Correlation Table

```
#url = "https://raw.githubusercontent.com/atfrank/CS-Annotate/master/data/train_target.csv"
url = "https://raw.githubusercontent.com/atfrank/CS-Annotate/Kexin/data/train_target_3.csv"
s=requests.get(url).content
parameter_data = pd.read_csv(io.StringIO(s.decode('utf-8')), sep = ' ')

corr = parameter_data.corr()
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr,cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(parameter_data.columns),1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(parameter_data.columns)
ax.set_yticklabels(parameter_data.columns)
plt.show()
```
